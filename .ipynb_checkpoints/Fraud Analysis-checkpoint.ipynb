```python
# Initial imports
import pandas as pd
import hvplot.pandas
from sqlalchemy import create_engine
```

```python
# create a connection to the database using the create_engine function
engine = create_engine("postgresql://postgres:postgres@localhost:5432/SQL Homework")
```

```python
# Define a query to pull purchases under $2 for holder IDs 18 and 2 from the holder_tx table
fraud_count_query = """select card_id, count(amount) 
        from holder_tx 
        where amount < 2 and (card_id = 18 or card_id = 2) 
        group by card_id 
        order by count(amount) desc"""

# Load data into the DataFrame using the read_sql() method from pandas
fraud_count_df = pd.read_sql(fraud_count_query, engine)
```

```python
fraud_count_df.head()
```

```python
fraud_history_query = """select card_id, amount, transaction_datetime
                   from holder_tx
                   where amount < 2 and (card_id = 18 or card_id = 2)
                   order by card_id"""

fraud_history_df = pd.read_sql(fraud_history_query, engine)

fraud_history_df
```

```python
# Slice data for each cardholder
fraud_history_2_df = fraud_history_df[:11]
fraud_history_18_df = fraud_history_df[-19:]
```

```python
# Plot transaction history

fraud_history_2_df.hvplot(x = 'transaction_datetime', y = 'amount')
```
